```python
#syntesing and over-sampling
#SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import cohen_kappa_score, make_scorer
def confusion_matrix_SMOTE(X_train,y_train,X_test,y_test):
    sm= SMOTE(random_state=2)
    X_train_SMOTE, y_train_SMOTE= sm.fit_resample(X_train, y_train)
#print(X_train_res.head)
#print(y_train_res.head)
    model.fit(X_train_SMOTE, y_train_SMOTE)
    print ('SMOTE result')
    y_pred_SMOTE=model.predict(X_test)
    print(y_pred_SMOTE) 
    conf_mat_SMOTE = confusion_matrix(y_true=y_test, y_pred=y_pred_SMOTE)
    print('Confusion matrix SMOTE:\n', conf_mat_SMOTE)
    kappa_SMOTE= cohen_kappa_score(y_test,y_pred_SMOTE, labels=None, weights='quadratic', sample_weight=None)
    print('kappa SMOTE')
    print(kappa_SMOTE)
    

confusion_matrix_SMOTE(X_train,y_train,X_test,y_test)
```

```python
#SMOTEENN - combine over&under samoling
from imblearn.combine import SMOTEENN

def confusion_matrix_SMOTEENN(X_train,y_train,X_test,y_test):
    sme= SMOTEENN(random_state=2)
    X_train_SMOTEENN, y_train_SMOTEENN= sme.fit_resample(X_train, y_train)
    model.fit(X_train_SMOTEENN, y_train_SMOTEENN)
    print ('SMOTEENN result')
    y_pred_SMOTEENN=model.predict(X_test)
    print(y_pred_SMOTEENN) 
    conf_mat_SMOTEENN = confusion_matrix(y_true=y_test, y_pred=y_pred_SMOTEENN)
    print('Confusion matrix SMOTEENN:\n', conf_mat_SMOTEENN)
    kappa_SMOTEENN= cohen_kappa_score(y_test,y_pred_SMOTEENN, labels=None, weights='quadratic', sample_weight=None)
    print('kappa SMOTEENN')
    print(kappa_SMOTEENN)
    

confusion_matrix_SMOTEENN(X_train,y_train,X_test,y_test)
```

```python
#Bagging
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
cart = DecisionTreeClassifier()
def confusion_matrix_Bagging(X_train,y_train,X_test,y_test):
    Bagging = BaggingClassifier(base_estimator=cart, n_estimators=100, random_state=2)
    Bagging.fit(X_train,y_train)
    y_pred_Bagging=Bagging.predict(X_test)
    conf_mat_Bagging = confusion_matrix(y_true=y_test, y_pred=y_pred_Bagging)
    print('Confusion matrix Bagging:\n', conf_mat_Bagging)
    kappa_Bagging= cohen_kappa_score(y_test,y_pred_Bagging, labels=None, weights='quadratic', sample_weight=None)
    print('kappa Bagging')
    print(kappa_Bagging)
    
confusion_matrix_Bagging(X_train,y_train,X_test,y_test)
```

```python
#Boosting
from sklearn.ensemble import GradientBoostingClassifier
Boosting=GradientBoostingClassifier(n_estimators=100, random_state=2)
def confusion_matrix_Boosting(X_train,y_train,X_test,y_test):
    Boosting.fit(X_train,y_train)
    y_pred_Boosting=Boosting.predict(X_test)
    conf_mat_Boosting = confusion_matrix(y_true=y_test, y_pred=y_pred_Boosting)
    print('Confusion matrix Boosting:\n', conf_mat_Boosting)
    kappa_Boosting= cohen_kappa_score(y_test,y_pred_Boosting, labels=None, weights='quadratic', sample_weight=None)
    print('kappa Boosting')
    print(kappa_Boosting)
    
print('Cross_validation')    
confusion_matrix_Boosting(X_train,y_train,X_test,y_test)
```
